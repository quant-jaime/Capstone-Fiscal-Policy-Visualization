```python
import numpy as np 
import pandas as pd 
import pandas_datareader.data as pdr
import matplotlib.pyplot as plt

import datetime as dt
%matplotlib inline
from matplotlib.pyplot import figure
```

```python
# s = setting t
# e = setting t+n
s = dt.datetime(1950, 1,1)
e = dt.datetime(2019,12,31)

# IMPGSC1 - imports of g/s 
# IEAXGS - exports of g/s 
# pandas_datareader.data collects collects financial data (FRED)
d = pdr.DataReader(['IMPGSC1', 'IEAXGS'], 'fred', s, e)

#plotting imports of g/s, enlarging the size if figsize 
d['IMPGSC1'].plot(figsize = (30,20))

#labeling the title,x/y axis 
plt.title('Imports of Goods & Services in USA')
plt.xlabel('Time')
plt.ylabel('Billions of Dollars')
# enlarging the font size and legend
plt.rc('figure', titlesize = 30)
```

```python
d['IEAXGS'].plot(figsize = (30,20))
plt.title('Exports of Goods & Services in USA')
plt.xlabel('Time')
plt.ylabel('Millions of Dollars')
plt.rc('figure', titlesize = 40)
```

```python
s1 = dt.datetime(1992,1,1)
e1 = dt.datetime(2019,12,31)

# BOPGSTB - trade balance, fred only has data since 1992
d1 = pdr.DataReader('BOPGSTB', 'fred', s1, e1)

d1.plot(figsize = (30,20))
plt.title('Trade Balance of USA')
plt.xlabel('Time')
plt.ylabel('Millions of Dollars')
plt.rc('figure', titlesize = 40)
```

```python
s2 = dt.datetime(1966, 1,1)
e2 = dt.datetime(2019,12,31)

d2 = pdr.DataReader('GFDEBTN', 'fred', s2, e2) 

d2.plot(figsize = (30,20))
plt.title('External Debt of USA')
plt.xlabel('Time')
plt.ylabel('Trillions of Dollars')
plt.rc('figure', titlesize = 40)
```
